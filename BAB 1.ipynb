```python
# Bab 1: Introduction
# Reproduksi kode dari buku "Introduction to Machine Learning with Python"

# Import library yang diperlukan
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

# **Mengapa Machine Learning?**
# Machine Learning membantu memecahkan masalah kompleks seperti deteksi wajah, klasifikasi email spam, dan lain-lain.

# **Dataset yang Digunakan**
# Dataset Iris adalah dataset klasik dalam pembelajaran mesin yang berisi informasi tentang tiga jenis bunga iris.

# Memuat dataset Iris
iris = load_iris()
print("Keys of iris_dataset: \n", iris.keys())
print("\nFeature names: \n", iris['feature_names'])
print("\nTarget names: \n", iris['target_names'])

# Menampilkan deskripsi dataset
print("\nDeskripsi dataset:\n", iris['DESCR'][:500], "...")

# Membagi data menjadi training dan testing
X_train, X_test, y_train, y_test = train_test_split(
    iris['data'], iris['target'], random_state=0
)

# Informasi tentang ukuran data
print("\nShape of X_train: ", X_train.shape)
print("Shape of y_train: ", y_train.shape)
print("Shape of X_test: ", X_test.shape)
print("Shape of y_test: ", y_test.shape)

# **Visualisasi Data**
# Membuat scatter plot data
iris_dataframe = pd.DataFrame(X_train, columns=iris['feature_names'])
scatter_matrix = pd.plotting.scatter_matrix(
    iris_dataframe, c=y_train, figsize=(15, 15), marker='o',
    hist_kwds={'bins': 20}, s=60, alpha=.8, cmap='viridis'
)
plt.suptitle("Scatter Matrix of Iris Dataset")
plt.show()

# **Membangun Model Machine Learning**
# Membuat model k-Nearest Neighbors
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train, y_train)

# Evaluasi model
accuracy = knn.score(X_test, y_test)
print("\nAkurasi pada data uji: {:.2f}".format(accuracy))

# **Prediksi dengan Data Baru**
# Prediksi dengan data baru
X_new = np.array([[5, 2.9, 1, 0.2]])
prediction = knn.predict(X_new)
print("\nData baru: ", X_new)
print("Prediction: ", prediction)
print("Predicted target name: ", iris['target_names'][prediction])

# **Kesimpulan**
# Model k-Nearest Neighbors berhasil dibuat dan diuji dengan akurasi tinggi pada data uji.
# Prediksi untuk data baru menunjukkan bahwa bunga tersebut kemungkinan besar adalah jenis 'setosa'.
```
