```python
# Bab 2: Supervised Learning
# Reproduksi kode dari buku "Introduction to Machine Learning with Python"

# Import library yang diperlukan
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_classification, make_regression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC

# **Supervised Learning: Konsep Dasar**
# Supervised Learning adalah jenis pembelajaran mesin di mana model dilatih menggunakan data berlabel.
# Terdapat dua jenis utama: klasifikasi (classification) dan regresi (regression).

# **1. Dataset untuk Klasifikasi**
X_class, y_class = make_classification(
    n_samples=100, n_features=2, n_informative=2, n_redundant=0,
    random_state=42, n_clusters_per_class=1
)

# Membagi data menjadi training dan testing
X_train_class, X_test_class, y_train_class, y_test_class = train_test_split(
    X_class, y_class, random_state=42
)

# Visualisasi dataset klasifikasi
plt.scatter(X_class[:, 0], X_class[:, 1], c=y_class, cmap='viridis', s=60)
plt.title("Dataset Klasifikasi")
plt.xlabel("Fitur 1")
plt.ylabel("Fitur 2")
plt.show()

# **2. Dataset untuk Regresi**
X_reg, y_reg = make_regression(
    n_samples=100, n_features=1, noise=10, random_state=42
)

# Membagi data menjadi training dan testing
X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(
    X_reg, y_reg, random_state=42
)

# Visualisasi dataset regresi
plt.scatter(X_reg, y_reg, color='blue')
plt.title("Dataset Regresi")
plt.xlabel("Fitur")
plt.ylabel("Target")
plt.show()

# **3. Model Klasifikasi**
# a. k-Nearest Neighbors
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train_class, y_train_class)
accuracy_knn = knn.score(X_test_class, y_test_class)
print("Akurasi k-NN: {:.2f}".format(accuracy_knn))

# b. Logistic Regression
logreg = LogisticRegression()
logreg.fit(X_train_class, y_train_class)
accuracy_logreg = logreg.score(X_test_class, y_test_class)
print("Akurasi Logistic Regression: {:.2f}".format(accuracy_logreg))

# c. Decision Tree
tree = DecisionTreeClassifier(max_depth=3, random_state=42)
tree.fit(X_train_class, y_train_class)
accuracy_tree = tree.score(X_test_class, y_test_class)
print("Akurasi Decision Tree: {:.2f}".format(accuracy_tree))

# d. Random Forest
forest = RandomForestClassifier(n_estimators=10, random_state=42)
forest.fit(X_train_class, y_train_class)
accuracy_forest = forest.score(X_test_class, y_test_class)
print("Akurasi Random Forest: {:.2f}".format(accuracy_forest))

# e. Support Vector Machine (SVM)
svm = SVC(kernel='linear')
svm.fit(X_train_class, y_train_class)
accuracy_svm = svm.score(X_test_class, y_test_class)
print("Akurasi SVM: {:.2f}".format(accuracy_svm))

# **4. Model Regresi**
# a. Linear Regression
linreg = LinearRegression()
linreg.fit(X_train_reg, y_train_reg)
score_linreg = linreg.score(X_test_reg, y_test_reg)
print("R^2 Linear Regression: {:.2f}".format(score_linreg))

# Visualisasi model regresi
plt.scatter(X_reg, y_reg, color='blue', label='Data')
plt.plot(X_reg, linreg.predict(X_reg), color='red', label='Model')
plt.title("Linear Regression")
plt.xlabel("Fitur")
plt.ylabel("Target")
plt.legend()
plt.show()

# **Kesimpulan**
# Bab ini menunjukkan beberapa algoritma supervised learning untuk klasifikasi dan regresi.
# Evaluasi model dilakukan menggunakan akurasi untuk klasifikasi dan R^2 untuk regresi.
```
